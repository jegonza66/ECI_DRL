## Cargar modelos desde torchvision
[`torchvision`](https://pytorch.org/docs/stable/torchvision/index.html) es un paquete con utilidades para implementar tareas de visión por computador basada en redes neuronales. Contiene numerosos modelos ya creados y entrenados, seleccionados entre los más exitosos de los últimos años en el desempeño de diversas tareas.

```python
from torchvision import models
dir(models)
```

Cargamos la red [AlexNet](http://papers.nips.cc/paper/4824-imagenet-classification-with-deep-convolutional-neural-networ) (2012), ya pre-entrenada para los retos de clasificación de imágenes con [ImageNet](http://www.image-net.org).

```python
alexnet = models.alexnet(pretrained=True)
alexnet
```

## Definir pre-procesamiento para la entrada
Definimos una transformación de los datos para ajustarlos al número de neuronas de entrada de la red.

```python
from torchvision import transforms
preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )])
```

## Calcular salida de la red sobre una imagen
Cargamos una imagen de prueba.

```python
from PIL import Image
import requests
from io import BytesIO

url = "https://github.com/jgromero/eci2019-DRL/blob/master/Tema%202%20-%20Aprendizaje%20Profundo/code/dragon1.jpg?raw=true"
# url = "http://sl.ugr.es/bobby_img"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

img
```

Preprocesamos la imagen para ajustarla a los requisitos de entrada de la red.

```python
import torch
img_t = preprocess(img)                # preprocesar imagen
batch_t = torch.unsqueeze(img_t, 0)    # convertir a tensor de 1-dimension
```

Calculamos la salida de la red para la imagen de entrada.

```python
alexnet.eval()          # entrar en modo evaluación
out = alexnet(batch_t)  # pasar la imagen ya preprocesada y ajustada
```

La salida es un array de 10000 entradas, cada una correspondiendo a una puntuación calculada por la red sobre la pertenencia de la imagen a una de las clases de [ImageNet](http://www.image-net.org). Comprobemos cuáles son las 3 clases a las que la red ha otorgado más puntuación.

```python
_, indices = torch.sort(out, descending=True)
indices[0][:3]
```

*Categorías de ImageNet* (ver [lista](https://gist.github.com/yrevar/942d3a0ac09ec9e5eb3a)):
- 207: 'golden retriever'
- 208: 'Labrador retriever'
- 852: 'tennis ball'
